```python
def rotate_image(list_2d):
    rotated_image = []
    for i in range(len(list_2d)):
        rotated_image.append([])
        for j in range(len(list_2d) - 1, -1, -1):
            rotated_image[i].append(list_2d[j][i])
    return rotated_image
```

```python
print(rotate_image([[5,1,9,11],[2,4,8,10],[13,3,6,7],[15,14,12,16]]))
```

```python
print(rotate_image([[1,2,3],[4,5,6],[7,8,9]]))
```

```python
print(rotate_image([[1,2],[3,4]]))
```
